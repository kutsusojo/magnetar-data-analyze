```python
from astropy import units as u
from astropy.coordinates import Galactic
from astropy.coordinates import SkyCoord
# from astropy_healpix import HEALPix
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from scipy.special import sph_harm as _sph_harm
import matplotlib.colors as clr
import pandas as pd
import time
# from tqdm.notebook import tqdm_notebook

# # カラーマップの設定
# cmap = {'red':((0.0,0.0,0.0),
#                  (0.25,0.0,0.0),
#                  (0.5,0,0),
#                  (0.75,0.9,0.9),
#                  (1.0,1.0,1.0)),
#           'green':((0.0,0.0,0.0),
#                    (0.25,0.75,0.75),
#                    (0.5, 1, 1),
#                    (0.75,0.9,0.9),
#                    (1.0,0.0,0.0)),
#           'blue':((0.0,1.0,1.0),
#                  (0.25,0.75,0.75),
#                  (0.5,0.,0.),
#                  (0.75,0.0,0.0),
#                  (1.0,0,0))
#           }
# mycolor = clr.LinearSegmentedColormap('mycolor', cmap)

# 変換定数
rad2deg = 180.0/np.pi
deg2rad = np.pi/180.0

# 球面調和関数
def sph_harm(l, m, theta, phi):
    return _sph_harm(m, l, phi, theta)

# sample a 360, 180 grid in RA/Dec
# 球面調和関数による展開を行うため 0.1°間隔でデータを取り出す必要がある。
# （1°間隔では、粗すぎて正しい展開係数が求まらない。）
factor = 10
ra = np.linspace(180., -180., 360*factor) * u.deg
dec = np.linspace(90., -90., 180*factor) * u.deg
ra_grid, dec_grid = np.meshgrid(ra, dec)

# set up astropy coordinate objects
coords = SkyCoord(ra_grid.ravel(), dec_grid.ravel(), frame='galactic')

# モルワイデ図法におけるパラメータ決定方程式
def equation(phi, theta):
    return 2*theta + np.sin(2*theta) - np.pi*np.sin(phi)

def f_theta(phi):

    if phi == 0.0:
        return 0.0

    x1 = -np.pi/2
    x2 = np.pi/2
    xm = (x1 + x2)/2
    err = 1e-12

    while(np.abs(x1 - x2) > err):
        xm = (x1 + x2)/2
        if equation(phi, x1)*equation(phi, xm) < 0:
            x2 = xm
        else:
            x1 = xm

    return xm

# モルワイデ図法での座標値の取得
lon_map = np.array([i for i in range(-180*factor, 180*factor)])*deg2rad/factor
lat_map = -np.array([i for i in range(-90*factor, 90*factor)])*deg2rad/factor
X = []
Y = []
for delta in lat_map:
    theta = f_theta(delta)
    for alpha in lon_map:
        x0 = 2*np.sqrt(2)*alpha/np.pi*np.cos(theta)
        y0 = np.sqrt(2)*np.sin(theta)
        X.append(x0)
        Y.append(y0)

# theta alpha of HS
theta_HS1 = np.radians(90 - 92.50)
alpha_HS1 = np.radians(123.45)
x_hs1 = 2*np.sqrt(2)*alpha_HS1/np.pi*np.cos(theta_HS1)
y_hs1 = np.sqrt(2)*np.sin(theta_HS1)
theta_HS2 = np.radians(90 - 159.97)
alpha_HS2 = np.radians(247.90 - 360)
x_hs2 = 2*np.sqrt(2)*alpha_HS2/np.pi*np.cos(theta_HS2)
y_hs2 = np.sqrt(2)*np.sin(theta_HS2)
# theta_HS3 = np.radians(90 - 31.3)
# alpha_HS3 = np.radians(122.2)
# x_hs3 = 2*np.sqrt(2)*alpha_HS3/np.pi*np.cos(theta_HS3)
# y_hs3 = np.sqrt(2)*np.sin(theta_HS3)

# グラフの設定
fig = plt.figure(figsize=(15, 6))

for i in range(1):
    ax = fig.add_subplot(1, 1, 1)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(False)
    plt.xticks(color="None")
    plt.yticks(color="None")
    plt.tick_params(length=0)
    plt.rcParams['font.family'] = 'IPAexGothic'  # <-- 日本語フォントを指定

    # 緯線の表示
    phi = np.array([i for i in range(-8, 9)])*10*deg2rad
    for p in phi:
        theta = f_theta(p)
        x = 2*np.sqrt(2)*np.array([i for i in range(-180, 180)])*deg2rad/np.pi*np.cos(theta)
        y = np.sqrt(2)*np.sin(theta)*np.array([1 for i in range(-180, 180)])
        ax.plot(x, y, color="gray", ls="--")

    # 経線の表示
    theta = np.array([i for i in range(-90, 91)])*deg2rad
    lon = np.array([i for i in range(-18, 19, 1)])*10
    for l in lon:
        x = 2*np.sqrt(2)*l*deg2rad/np.pi*np.cos(theta)
        y = np.sqrt(2)*np.sin(theta)
        ax.plot(x, y, color="gray", ls="--")

#     # 温度ゆらぎ分布の表示
#     sc = ax.scatter(X, Y, c=tmap_gauss.flatten(), marker='o', s=10, edgecolor='',
#                     cmap=mycolor, vmin=-0.2, vmax=0.2)
#     plt.colorbar(sc)
ax.scatter(x_hs1, y_hs1, c="blue")
ax.scatter(x_hs2, y_hs2, c="orange")
# ax.scatter(x_hs3, y_hs3, c="red")

# plt.savefig("Fig_02.png")
plt.show()
```

```python
# %matplotlib qt
# spectrum, using SGR 0418+5729 fitting results

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation
import numpy as np
import math
import scipy.optimize
import os
import matplotlib
from mpl_toolkits.mplot3d import axes3d, Axes3D

# physics constant
c = 3 * 10 ** 10
k_B = 8.617333262 * 10 ** (-5) # eV K-1, Boltzmann constant
h = 4.135667696 * 10 ** (-15) # eV⋅Hz−1, Planck constant

# geometric input
# inclination angle
# i = np.radians(54.59229606284669)
# hot spot parameter
# theta_1 = np.radians(170.30032148969232)
# phi_1 = np.radians(128.15356622008315)
norm_1 = 6.462363930124616
# theta_2 = np.radians(172.0399266405045)
# phi_2 = np.radians(251.3109205065553)
norm_2 = 18.615967112346844
# theta_3 = np.radians(31.2795000977478)
# phi_3 = np.radians(122.16917843409833)
norm_3 = 1.2833967237907484

T_1 = norm_1 ** (1 / 4)
T_2 = norm_2 ** (1 / 4)
T_3 = norm_3 ** (1 / 4)

# constant parameter (eV units)
c = 2.998 * 10 ** 10
D_10 = 6 * 10 ** 21 # D_10 = distance / 10kpc
R_magnetar = 10 ** 6
k_B = 8.617 * 10 ** (-5)
K = R_magnetar ** 2 / D_10 ** 2 # normalization factor

# planck distribution
# def spectrum(x, T):
#     return (2 * x ** 3) / ((np.e ** (x / (k_B * T)) - 1) * c ** 2 * h ** 2)

# planck distribution in X-ray era
def spectrum(x, T):
    return (K * 1.0344 * 10 ** (-3) * x ** 2) / (np.e ** (x / k_B / T) - 1)


E = np.logspace(0, 1000000, 10000)
# E = h * nu (old version)

spec_HS1 = spectrum(E, T_1)
spec_HS2 = spectrum(E, T_2)
spec_HS3 = spectrum(E, T_3)

fig = plt.figure(figsize = (12.0, 6.0), dpi = 300)
plt.plot(E, spec_HS1, color = "pink")
plt.plot(E, spec_HS2, color = "red")
plt.plot(E, spec_HS3, color = "blue")
# plt.xlim(0, 1000)
# plt.ylim(0, 100)
plt.show()
```
